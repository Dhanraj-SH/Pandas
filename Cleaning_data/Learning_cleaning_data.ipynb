# PANDAS - CLEANING DATA

#### Data Cleaning

Data cleaning means fixing bad data in your data set

Bad Data could be:

- Empty cells
- Data in wrong format
- Wrong Data
- Dulplicates


#### Cleaning empty cells


```python
'''
Empty Cells -

Empty cells can potentially give you a wrong result when you analyze the data


Remove Rows -

One way to deal with empty cells is to remove rows that contain empty cells.
This is usually OK, since data sets can be very big, and removing a few rows will not have a big impact on the result.
'''

import pandas as pd

df = pd.read_csv('data.csv')
dup_df = df.dropna() #by default dropna() method returns a new DataFrame, and will not change the original

dup_df 
```

```python

#before changing the the original DataFrame
df = pd.read_csv('data.csv')
df 
```

```python
#if we want to change the Orignal DataFrame, use the inplace=True argument
df.dropna(inplace= True)

#after  changing the original DataFrame
df
```

#### Replace Empty Values

Another way of dealing with empty cells is to insert a new value instead.

This way you do not have to delete entire rows just because of some empty cells.

The fillna() method allows us to replace empty cells with values

```python
df = pd.read_csv('data.csv')

df.fillna(130, inplace = True)

df
```

```python
'''
Replace only for Specifed Colums

The example above replaces all empty cells in the whole Data Frame.
To only replace empty values for one column, specify the column name for the DataFrame
'''

df = pd.read_csv('data.csv')

df.fillna({"Calories": 130}, inplace = True)

df
```

```python
'''
Replacing using mean, median or mode

A common way to replace empty cells, is to calculate the mean, median and mode value of the column.

Pandas uses the mean(), median(), and mode() methods to calculate the respective values fo a specified column
'''

df = pd.read_csv('data.csv')
x = df["Calories"].mean()
df.fillna({"Calories": x}, inplace=True)

df
#MEAN = the avg value(the sum of all divided by number of values)
```

```python
df = pd.read_csv('data.csv')
x = df["Calories"].median()
df.fillna({"Calories":x}, inplace=True)

df

#MEDIAN = The value in the middle, after you have sorted all values ascending
```

```python
df = pd.read_csv('data.csv')
x = df["Calories"].mode()
df.fillna({"Calories": x}, inplace= True)

df

#MODE = The value that appears most frequently.
```

####    Pandas - Cleaning data of wrong format

Data of wrong format

Cells with data of wrong format can make it difficult, or even impossible, to analyze data.

To fix it, you have two options
- Remove the rows
- Or convert all cells in the columns into the same format

```python
# Convert into a correct format
'''
df = pd.read_csv('data.csv')
df['Date'] = pd.to_datetime(df['Date'], format='mixed') #this line of code can only be exectued for the pandas above 2.2.x version

df
'''

# so will be using another approcah for lower version of pandas
df = pd.read_csv('data.csv')
df['Date'] = df['Date'].str.replace("'", "")
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

df
```

```python
#As the result of conveting in above dataset it gave us a NaT value, which can be handled as a NULL value.
#And we can remove it by using dropna() method

df.dropna(subset=['Date'], inplace = True)
```

#### FIXING WRONG DATA

Wrong Data

"Wrong data" does not have to be "empty cells" or "wrong format". it can just be wrong, like if someone registed "199" instead of 1.99.

Sometimes you can spot wrong data by looking at the data set, because u have a exception of what it should it be

If you take a look at our data set, you can see that in row 7, the duration is 450, but for all the other rows the duration is between 30 to 60.

It doesn't have to be wrong, but taking in consideration that this is the data set of someone's workout sessions, we conclude with the fact that this person did not workout in 450 minutes

```python
'''
Replacing values

One way to fix wrong values is to replace them with somthing else.
In example its most likely a typo, and value should be "45" instead of "450", and we coulf just instead "45"
 in row 7
'''

df.loc[7,'Duration'] = 45


'''
For small data sets you might be able to replace the wrong data one by one, but not for big data sets.

To replace wrong data for larger data sets you can create some rules, e.g. set some boundaries for legal values, and replace any values that are outside of the boundaries.
'''

for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.loc[x, "Duration"] = 120


'''
Removing rows

Another way of handling wrong data is to remove the rows that contains wrong data.
This way you do not have to find out what to replace them with, and there is a good chance you do not need them to do your analyses.
'''


for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.drop(x, inplace = True)
```

#### REMOVING DUPLICATES

Discovering Duplicates - Duplicates rows are rows that have been registered more than one time.

To discover duplicates, we can use the `duplicated()` method.

The `duplicated()` method returns a Boolean values for each row:

```python
df.duplicated() # retuens ture for every row that has been duplicated
```

```python
#Removing duplicates - to remove duplicates use the drop_duplicates method

df.drop_duplicates(inplace = True)
```
